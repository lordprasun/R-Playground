```python
import numpy as np
import pandas as pd
```

```python
series_obj = pd.Series(data=[100,200,300,np.nan,500,600,np.nan,800,900,1000],name='Check Null')
series_obj.isnull()
```

```python
DF_obj = pd.DataFrame(np.random.rand(36).reshape(6,6))
DF_obj.loc[2:4,5]=np.nan
DF_obj.loc[5,2:4]=np.nan
DF_obj
```

```python
filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
filled_DF = DF_obj.fillna({0: 0.1, 5:1.25})
filled_DF
```

```python
filled_DF = DF_obj.fillna(method='ffill')
filled_DF
```

```python
DF_no_NaN = DF_obj.dropna()
DF_no_NaN
```

```python
DF_no_NaN = DF_obj.dropna(axis=1)
DF_no_NaN
```
