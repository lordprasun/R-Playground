```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,4,5])
ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```

```python
x = np.linspace(0, 2*np.pi, 400)
fig = plt.figure()
fig, (ax, ay) = plt.subplots(2,2)

ax[0].plot(x)
ax[1].plot(x,np.sin(x**2))
ay[0].plot(x,np.cos(x**2))
ay[1].plot(x,np.tan(x**2))
```
